```python
from snownlp import SnowNLP
import pandas as pd
import jieba.analyse
```

```python
df = pd.read_excel("D:/微博数据汇总1.xlsx")
df['tag'] = None
df.head
```

```python
for i in range(1,len(df)):
    try:
        s = SnowNLP(''.join(jieba.analyse.textrank(df.loc[i]['text'])))
    except:
        s = SnowNLP(df.loc[i]['text'])
    
    if s.sentiments <=0.2:
        tag = -2
    elif 0.2<s.sentiments<=0.4:
        tag = -1
    elif 0.4<s.sentiments<=0.6:
        tag = 0
    elif 0.6<s.sentiments<=0.8:
        tag = 1
    elif 0.8<s.sentiments:
        tag = 2
    df.loc[i,'tag'] = tag
    print(i,tag)
```

```python
df.to_excel("D:/微博情感分析后.xlsx")
```
